```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

#### Method 1

```python
plt.plot(x, y); # semicolon to omit other outputs
```

#### Method 2: Object Oriented Approach (much more flexible)

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
plt.style.use("seaborn")
```

```python
y = np.exp(x)
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
ax[0].plot(x, y)
ax[1].scatter(x, y, marker="+", s=25); # "+" marker, size of marker
```

```python
# plotting from a dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter": 12
}
fig, ax = plt.subplots(2, 1)
ax[0].set(
    title="Nut butter prices",
    xlabel="Nut butter",
    ylabel="Prices",
)
ax[0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1].set(
    title="Nut butter prices",
    xlabel="Nut butter",
    ylabel="Prices",
)
ax[1].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
fig.tight_layout()
```

```python
x = np.random.randn(1000)
plt.hist(x);
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 5))
ax[0,0].plot(x, x)
ax[0,1].hist(x)
ax[1,0].scatter(np.random.random(10), np.random.random(10))
ax[1,1].scatter(x, np.tanh(x))
fig.tight_layout()
plt.show()
print(ax)
```
